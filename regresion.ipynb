## Ejemplo de regresión lineal

```python
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import numpy as np
```

```python
#x es el dinero invertido, y es el dinero ganado
#el caso lineal se plantea como una recta a partir de los puntos (x,y)
x = np.array([0,6.5,13.2,18.1,24.9,28.2,36.1])
y = np.array ([5,10,17,20,25,30,35])

plt.figure(figsize=(7,7))
plt.scatter(x, y, s=60)
plt.ylim(bottom=0)
plt.title("Valores")
plt.xlabel("Dinero Invertido")
plt.ylabel("Dinero Ganado")
plt.grid()
plt.show()
```

```python
#se busca una función acorde al gráfico, esto es la regresión lineal
a = np.arange(0,40)
b = 0.8 * a + 5

plt.figure(figsize=(7,7))
plt.plot(a, b, color="red", label="Nuestra solución")
plt.scatter(x, y, s=60)
plt.ylim(bottom=0)
plt.title("Valores")
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.grid()
plt.show()
```
